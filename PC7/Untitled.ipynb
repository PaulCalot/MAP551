```python
# Question 2.2.2

import numpy as np
def phi(x):
    alpha = np.sqrt(1-np.exp(-x))
    return 0.5*np.exp(-x/2)*np.log((1+alpha)/(1-alpha))
```

```python
# plotting the function
import matplotlib.pyplot as plt

X = np.linspace(0,5,2000)
Y = phi(X)


fig, ax = plt.subplots(figsize=(15,10))
plt.title("Graph of $\Psi(x) = e^{-x/2}cosh^{-1}(e^{x/2})$")
plt.plot(X,Y)
print(np.max(Y))
plt.savefig("phi_plot_2_2_2.png", dpi = 300)
```
